# Problem 1.6 : Convergence of Forth order Finite Difference

Let us compute the derivative for two functions as follows:
$$
u(x) = exp(sin^2(x))
$$
$$
u(x) = exp(sin(x)\vert(sin(x)\vert)
$$
using fourth order finite difference scheme
$$
u'(x_j) \approx w_j = \frac{1}{h} \left( \frac{1}{12} u_{j-2} - \frac{2}{3} u_{j-1} + \frac{2}{3} u_{j+1} - \frac{1}{12} u_{j+2} \right)
$$
using periodic boundary conditions.

```python
%matplotlib inline
from scipy.sparse import coo_matrix
from numpy import arange,pi,exp,sin,cos,ones,sign
from numpy.linalg import norm
from math import factorial
from matplotlib.pyplot import figure,loglog,hold,semilogy,text,grid,xlabel,ylabel,title
import time
```

Computing for $$
u(x) = exp(sin^2(x))
$$

```python
Nvec = 2**arange(3,17)

figure(figsize(5,5))
  
count = 0
a = np.zeros(len(Nvec))

for N in Nvec:
    t0 = time.time()
    h = 2*pi/N
    x = -pi + arange(1,N+1)*h
    u = exp(sin(x))
    # uprime is the 1st derivative
    uprime = cos(x)*u
    u_5prime = (15*sin(x)**2 - 10*sin(x)*cos(x)**2 + 15*sin(x) + cos(x)**4 - 10*cos(x)**2 + 1)*exp(sin(x))*cos(x)
    leading_e = norm((-(4.0)/factorial(5.0))*(h**4)*(u_5prime),inf)
    e = ones(N)
    e1 = arange(0,N)
    e2 = arange(1,N+1)
    e2[N-1]=0
    e3 = arange(2,N+2)
    e3[N-2]=0; e3[N-1]=1;
    D = coo_matrix((2*e/3,(e1,e2)),shape=(N,N))
    D = D - coo_matrix((e/12,(e1,e3)),shape=(N,N))
    D = (D - D.T)/h
    error = norm(D.dot(u)-uprime,inf)
    t1 = time.time()
    figure(1)
    loglog(N,error,'or',N,leading_e,"*")
    #hold(True)
    #print abs(error- leading_e)
    figure(2)
    loglog(N,abs(error - leading_e),'--')
    #hold(True)
    totaltime = t1-t0
    a[count] = totaltime
    count += 1

figure(1)
semilogy(Nvec,Nvec**(-4.0),'--')
text(105,5e-8,'$N^{-4}$')
grid(True)
xlabel('N')
ylabel('error')
title('Convergence of fourth-order finite difference');
```

Computing for $$
u(x) = exp(sin(x)\vert(sin(x))\vert)
$$

```python
Nvec = 2**arange(3,13)

figure(figsize(5,5))
  
count = 0
a = np.zeros(len(Nvec))

for N in Nvec:
    t0 = time.time()
    h = 2*pi/N
    x = -pi + arange(1,N+1)*h
    u = exp(sin(x)*abs(sin(x)))
    # uprime is the 1st derivative
    uprime = sign(x)*(2*exp(sin(x)*abs(sin(x)))*sin(x)*cos(x))
    e = ones(N)
    e1 = arange(0,N)
    e2 = arange(1,N+1)
    e2[N-1]=0
    e3 = arange(2,N+2)
    e3[N-2]=0; e3[N-1]=1;
    D = coo_matrix((2*e/3,(e1,e2)),shape=(N,N))
    D = D - coo_matrix((e/12,(e1,e3)),shape=(N,N))
    D = (D - D.T)/h
    error = norm(D.dot(u)-uprime,inf)
    t1 = time.time()
    loglog(N,error,'or')
    hold(True)
    totaltime = t1-t0
    a[count] = totaltime
    count += 1

semilogy(Nvec,Nvec**(-1.0),'--')
text(100,5e-3,'$N^{-1}$')
grid(True)
xlabel('N')
ylabel('error')
title('Convergence of fourth-order finite difference');
```
