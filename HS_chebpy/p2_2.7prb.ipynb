# Chapter 2 problem 2.7: 

Modify Program 3 to determine the maximum error over 1R in the sine function
interpolants of the square wave and the hat function, and to produce a log-log plot of
these two error maxima as functions of h. Good choices for h are $$2^{-3} , 2^{-4} , 2^{-5} , 2^{-6}$$
What convergence rates do you observe? as $$h\rightarrow 0$$

Program : 3 
Given in the text book will interpolate following functions using band limited interpolation on an infinite grid.

Delta function
$$
v(x) = \begin{cases}
1 & x =0 \\
0 & otherwise
\end{cases}
$$

Square wave
$$
v(x) = \begin{cases}
1 & |x| \le 3 \\
0 & otherwise
\end{cases}
$$

Hat function
$$
v(x) = \max(0, 1-|x|/3)
$$

Since all functions are zero away from origin, we can restrict to some finite interval, say $[-10,10]$.

```python
from numpy import arange,maximum,abs,zeros,sin,pi
from matplotlib.pyplot import subplot,figure,plot,grid,axis,loglog,semilogy
```

```python
h = 1/(2.0**(np.arange(3,7))); 
#h = 1/(2.0**(-3))
xmax = 10.0;
e_comp = zeros((3,len(h)))
for k in range(len(h)):
    x = arange(-xmax,xmax+h[k],h[k])
    xx = arange(-xmax-h[k]/20, xmax+h[k]/20, h[k]/10)
    
    figure(figsize(10,10))
    for pl in range(3):
        subplot(4,1,pl+1)
        if pl==0:
            v = (x==0)                      # delta function
        elif pl==1:
            v = (abs(x) <= 3.0)             # square wave
        elif pl==2:
            v = maximum(0.0,1.0-abs(x)/3.0) # hat function
        plot(x,v,'.')
        grid(True)
        p = zeros(len(xx))
        error = zeros(len(x))
        for i in range(len(x)):
            p = p + v[i]*sin(pi*(xx-x[i])/h[k])/(pi*(xx-x[i])/h[k])
            error[i] = (p[i]-v[i])
        e_comp[pl][k] = max(error)
        plot(xx,p)
        axis([-xmax,xmax,-0.5,1.5]);

figure(1)
figure(figsize(5,5))
xlabel("Number of Grid Points")
ylabel("Error")
#plotting error for only Square wave & Hat Function
for q in range(1,3):
    loglog(h,e_comp[q,:],'*')
    hold(True)
semilogy(h,h**(1),"--")
text(15e-3,5e-2,'$N^1$')
semilogy(h,h**(2),"--")
text(15e-3,5e-5,'$N^2$')
```
