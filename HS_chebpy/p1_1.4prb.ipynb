# Chapter 1 Exercise Problem 1.4: Convergence of fourth order finite differences

Let us compute the derivative of
$$
u(x) = \exp(\sin(x)), \qquad x \in [-\pi,\pi]
$$
using fourth order finite difference scheme
$$
u'(x_j) \approx w_j = \frac{1}{h} \left( \frac{1}{12} u_{j-2} - \frac{2}{3} u_{j-1} + \frac{2}{3} u_{j+1} - \frac{1}{12} u_{j+2} \right)
$$
using periodic boundary conditions.

```python
%matplotlib inline
from scipy.sparse import coo_matrix
from numpy import arange,pi,exp,sin,cos,ones
from numpy.linalg import norm
from matplotlib.pyplot import figure,loglog,hold,semilogy,text,grid,xlabel,ylabel,title
import time
```

```python
Nvec = 2**arange(3,17)

figure(figsize(5,5))
  
count = 0
a = np.zeros(len(Nvec))

for N in Nvec:
    t0 = time.time()
    h = 2*pi/N
    x = -pi + arange(1,N+1)*h
    u = exp(sin(x))
    uprime = cos(x)*u
    e = ones(N)
    e1 = arange(0,N)
    e2 = arange(1,N+1)
    e2[N-1]=0
    e3 = arange(2,N+2)
    e3[N-2]=0; e3[N-1]=1;
    D = coo_matrix((2*e/3,(e1,e2)),shape=(N,N))
    D = D - coo_matrix((e/12,(e1,e3)),shape=(N,N))
    D = (D - D.T)/h
    error = norm(D.dot(u)-uprime,inf)
    t1 = time.time()
    loglog(N,error,'or')
    hold(True)
    totaltime = t1-t0
    a[count] = totaltime
    count += 1

semilogy(Nvec,Nvec**(-4.0),'--')
text(105,5e-8,'$N^{-4}$')
grid(True)
xlabel('N')
ylabel('error')
title('Convergence of fourth-order finite difference');
```
